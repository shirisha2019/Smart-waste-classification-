```python
import os
from PIL import Image
```

```python
dataset_path = r"C:\Users\Thirupathiamma\sustainability\archive\DATASET"
```

```python
for root, dirs, files in os.walk(dataset_path):
    print(root, "contains", len(files), "files")
```

```python
bad_images = []

for root, dirs, files in os.walk(dataset_path):
    for file in files:
        if file.lower().endswith(('.png', '.jpg', '.jpeg')):
            file_path = os.path.join(root, file)
            try:
                img = Image.open(file_path)
                img.verify()
            except Exception as e:
                print("‚ùå Bad file:", file_path)
                bad_images.append(file_path)

print(f"‚úÖ Total bad images found: {len(bad_images)}")
```

```python
import hashlib

def find_duplicates(folder):
    hashes = {}
    duplicates = []
    for root, dirs, files in os.walk(folder):
        for file in files:
            file_path = os.path.join(root, file)
            with open(file_path, 'rb') as f:
                file_hash = hashlib.md5(f.read()).hexdigest()
            if file_hash in hashes:
                duplicates.append(file_path)
            else:
                hashes[file_hash] = file_path
    return duplicates

dupes = find_duplicates(dataset_path)
print(f"üîÅ Total duplicate images found: {len(dupes)}")

# (Optional) Remove them:
for d in dupes:
    os.remove(d)
print("‚úÖ Duplicates removed successfully!")
```

```python
extensions = set()
for root, dirs, files in os.walk(dataset_path):
    for file in files:
        if '.' in file:
            ext = file.split('.')[-1].lower()
            extensions.add(ext)
print("üîç File extensions found:", extensions)
```

```python
added = 0
for root, dirs, files in os.walk(dataset_path):
    for file in files:
        old_path = os.path.join(root, file)
        if os.path.isfile(old_path) and '.' not in file:
            new_path = old_path + '.jpg'
            os.rename(old_path, new_path)
            added += 1
print(f"‚úÖ Total files renamed (added .jpg): {added}")

```
